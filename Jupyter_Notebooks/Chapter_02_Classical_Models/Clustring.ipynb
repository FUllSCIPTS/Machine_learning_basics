```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.mixture import GaussianMixture
X_train = np.array(data)
```

```python
cov_types = ["spherical", "diag", "tied", "full"]

for c_t in cov_types:
    gmm = GaussianMixture(n_components=5,covariance_type=c_t)
    gmm.fit(X_train)

    print('\033[1m' + c_t + '\033[0m')
    if c_t=="full":
        print("each component has its own general covariance matrix")
    elif c_t=="tied":
        print("all components share the same general covariance matrix")
    elif c_t=="diag":
        print ("each component has its own diagonal covariance matrix")
    else:
        print("each component has its own single variance")

    print("cov matrix:",gmm.covariances_)

    X, Y = np.meshgrid(np.linspace(-40, 40), np.linspace(-40,40))
    XX = np.array([X.ravel(), Y.ravel()]).T
    Z = gmm.score_samples(XX)
    Z = Z.reshape((50,50))
    plt.contour(X, Y, Z)
    plt.scatter(X_train[:, 0], X_train[:, 1])
    plt.show()
    print("-------------------------------------------------------------------")
```
